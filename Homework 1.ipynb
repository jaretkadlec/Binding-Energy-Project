```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Exercise 2.2a
# The equation for height above the surface of the Earth for geosyncronous orbits is h = (GMT^2/4pi^2)^1/3 - R
# To get to this equation we start off with the eqation for period T = 2pi*r/v 
# and balancing the forces mv^2/r = GmM/r^2 (Centripital force on the right and gravitational on the left)
# Switch around the period equation v = 2pi*r/T and plug it into the force equation we found (2pi*r/T)^2 = GM/r
# Doing some algebra we get 
# r^3 = GMT^2/r*pi^2
# r = (GMT^2/r*pi^2)^1/3
# and the radius of the orbit is the radius of the Earth 
# plus the height above the surface of the Earth r = R + h
# Exercise 2.2c

G = 6.67*10**-11
M = 5.97*10**24
R = 6371*10**3 #meters
T = 86400 #seconds
#Calculating the height above the surface of the Earth
h = ((G*M*T**2)/(4*math.pi**2))**(1/3) - R
print('The height of the orbit for 86400 seconds is', '%.2f' % h, 'meters')

G = 6.67*10**-11
M = 5.97*10**24
R = 6371*10**3 #meters
T = 5400 #seconds
h = ((G*M*T**2)/(4*math.pi**2))**(1/3) - R
print('The height of the orbit for 5400 seconds is', '%.2f' % h, 'meters')

G = 6.67*10**-11
M = 5.97*10**24
R = 6371*10**3 #meters
T = 2700 #seconds
h = ((G*M*T**2)/(4*math.pi**2))**(1/3) - R
print('The height of the orbit for 2700 seconds is', '%.2f' % h, 'meters')
# The last of these calculations is not possible, since 
# it's lower than the surface of the Earth
# Exercise 2.2d
# For a solar day, since the sun has orbited through space, 
# it must rotate a bit more to be in the position it was in in the past day. 
# Therefore, the sidereal day is a bit shorter than a solar day.
```
